```julia
using Julog, Test
```

```julia
fib_rules = @julog [
    fib(0, 0) <<= true,
    fib(1, 1) <<= true,
    fib(X, F) <<= (X >= 2) & fib(X - 1, F1) & fib(X - 2, F2) & is(F, F1 + F2),
];
```

```julia
sat, subst = resolve(@julog(fib(7, F)), fib_rules)
```

```julia
@test @varsub({F => 13}) in subst
```
