```julia
using Julog, Test
```

```julia

demand_rules = @julog [
    demand_fib(N1) <<= demand_fib(N) & (N >= 1) & is(N1, N - 1),
    demand_fib(N2) <<= demand_fib(N) & (N >= 2) & is(N2, N - 2),
];
```

```julia
demand_fib7 = @julog [
    demand_fib(7) <<= true
]
```

```julia
demands = derivations(vcat(demand_rules, demand_fib7), Inf)
```

```julia
derive(demand_fib7, demand_rules; max_steps=20)[2]
```

```julia
fib_rules = @julog [
    fib(0, 0) <<= demand_fib(0),
    fib(1, 1) <<= demand_fib(1),
    fib(X, F) <<= demand_fib(X) & (X >= 2) &
        fib(X - 1, F1) & fib(X - 2, F2) & is(F, F1 + F2),
];
```

```julia
fib_facts = derivations(fib_rules, demands, Inf)
```

```julia
goals = @julog [fib(7, F)]
sat, subst = resolve(goals, fib_facts)
```

```julia
@test @varsub({F => 13}) in subst
```
